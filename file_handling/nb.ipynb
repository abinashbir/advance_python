### The moment we call the open funtion it loads into the RAM so it is neccesary to close the file 

# Single line add


```python
file = open("sample.txt","w")

file.write("heloooooooo")

file.close()
```

# Multiple line add

```python
file=open("sample2.txt",'w')
file.write('heloo')
file.write('\nhow are you')
file.close()
```

# Add line to a existing file
### This removes everything in the file

```python
f = open('sample.txt','w')
f.write('asdfg')
f.close()
```

### Appending lines without removing the existing content.

```python
f = open('sample2.txt','a')
f.write('I have appended some thing..')
f.close()
```

### Appending multiple lines

```python
#list of lines

l=['hi\n','I am\n','a programmer']

f=open('/home/abinash/adv_python/file_handling/sample.txt','w')
f.writelines(l)
f.close()
```

# Reading a file

### Reading the whole file


```python
f = open('/home/abinash/adv_python/file_handling/sample.txt','r')

s=f.read()
print(s)
f.close()
```

### Reading by given number

```python
f = open('/home/abinash/adv_python/file_handling/sample.txt','r')

s=f.read(6)
print(s)
f.close()
```

### Reading line-by-line




```python
f = open('/home/abinash/adv_python/file_handling/sample.txt','r')
print(f.readline())
print(f.readline())
print(f.readline())
f.close()
```

### Reading line-by-line with loop

```python
f = open('/home/abinash/adv_python/file_handling/sample.txt','r')

lines=f.readlines()

for i in lines:
    print(i)
```
