### Using with keyword as it automatically closes the files...

```python
# Write

with open('/home/abinash/adv_python/file_handling/sample2.txt','w') as f:
    f.write('yayyyyyyyyyyyyyy')
```

```python
# Read

with open('/home/abinash/adv_python/file_handling/sample2.txt','r') as f:
    print(f.read())
```

```python
# Reading the characters with respect to number f.read(5)

with open('/home/abinash/adv_python/file_handling/sample2.txt','r') as f:
    print(f.read(5))
```

### Loading a big file in the buffer register


```python
big_list =['hello world' for i in range(1000)]

with open('big.txt','w') as f:
    f.writelines(big_list)
```

### Loading a big file in chunks

```python
with open('big.txt','r') as f:
    chunk_size=100

    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size))
        f.read(chunk_size)
```

### Printing the next index of the character in a txt file

```python
with open('sample.txt','r') as f:
    print(f.read(5))
    print(f.tell())
    
```

### Reloacting the buffer index in a txt file using the seek()

```python
# Seef have power to teleport you to any index within the lenth of the txt file

with open('sample.txt','r') as f:
    print(f.read(5))
    print(f.tell())
    f.seek(0)

    print(f.read(5))
```
