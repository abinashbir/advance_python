# dump() -> create a json file
# load() -> print or load a json file

### Converting list to json


```python
import json

l=[1,2,3,4]

with open('demo.json','w') as f:
    json.dump(l,f)
```

### Converting dict to json


```python
dict={
    'name':'abinash',
    'age':21,
    'gender':'male'
}

with open('demo.json','w') as f:
    json.dump(dict,f,indent=4) #the indent argument is used to show the json in a good readable manner
```

### Desirialization

```python
import json

with open('demo.json','r') as f:
    d=json.load(f)
    print(d)
    print(type(d))
```
